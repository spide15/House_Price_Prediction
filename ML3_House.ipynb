```python
import pandas as pd
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

```python
from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()
```

```python
housing
```

```python
df=pd.DataFrame(housing.data,columns=housing.feature_names)
```

```python
df.head()
```

```python
df['Price']=housing.target
```

```python
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
correlation=df.corr()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(correlation,cbar=True,square=True,fmt=".2f",annot=True,annot_kws={'size':8},cmap="Blues")
```

```python
X=df.drop(['Price'],axis=1)
```

```python
Y=df['Price']
```

```python
X
```

```python
Y
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=1)
```

```python
print(X.shape,x_train.shape,x_test.shape)
```

```python
model=XGBRegressor()
```

```python
model.fit(x_train,y_train)
```

```python
training_data_accuracy=model.predict(x_train)
```

```python
print(training_data_accuracy)
```

```python
# Rsquare
```

```python
score_1=metrics.r2_score(y_train,training_data_accuracy)
```

```python
print(score_1)
```

```python
score_2=metrics.mean_absolute_error(y_train,training_data_accuracy)
```

```python
print(score_2)
```

```python
# testing_data
```

```python
test_data_prediction=model.predict(x_test)
```

```python
score_11=metrics.r2_score(y_test,test_data_prediction)
```

```python
print(score_11)
```

```python
score_22=metrics.mean_absolute_error(y_test,test_data_prediction)
```

```python
print(score_22)
```

```python
plt.scatter(y_train, training_data_accuracy)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Price vs Preicted Price")
plt.show()
```

```python
plt.scatter(y_test, test_data_prediction)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Price vs Preicted Price")
plt.show()
```
